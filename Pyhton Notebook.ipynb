### c) Filtering Data for Countries with Multiple Participants

```python

if df_all is not None:
    #  Get the count of participants per country
    country_counts = df_all['Country'].value_counts()

    # Identify countries where the count is greater than 1
    countries_to_keep = country_counts[country_counts > 1].index

    # Filter the original DataFrame, Keep only the rows where the 'Country' is in the list of countries_to_keep
    df_filtered = df_all[df_all['Country'].isin(countries_to_keep)].copy()
    
    print("Filtered DataFrame (df_filtered) created.")

    # Get the number of rows in the filtered DataFrame
    num_rows_filtered = df_filtered.shape[0] # Or use len(df_filtered)

    print(f"\nNumber of rows in the filtered DataFrame (df_filtered): {num_rows_filtered}")

    # Print the original number of rows for comparison
    print(f"Original number of rows in df_all: {df_all.shape[0]}")

    # Display the first few rows of the filtered DataFrame
    print("\nFirst 5 rows of df_filtered:")
    display(df_filtered.head())
```

### d) Average Daily Usage Hours Per Country from Flitered Data

```python

if 'df_filtered' in locals() and df_filtered is not None:
    if not df_filtered.empty:
        # Calculate the average daily usage hours per country, Group by 'Country' and calculate the mean of 'Avg_Daily_Usage_Hours'
        average_usage_per_country = df_filtered.groupby('Country')['Avg_Daily_Usage_Hours'].mean()

        # Find the top ten countries with the highest average usage, Sort the results in descending order and take the top 10
        top_10_avg_usage = average_usage_per_country.sort_values(ascending=False).head(10)

        print("Top 10 Countries by Average Daily Usage Hours:")
        print(top_10_avg_usage)

        # --- Visualize the top ten countries by average usage ---
        plt.figure(figsize=(12, 7)) # Adjust figure size as needed

        # Create a bar plot
        sns.barplot(x=top_10_avg_usage.index, y=top_10_avg_usage.values,hue=top_10_avg_usage.index, palette='viridis',legend=False)

        plt.title('Top 10 Countries by Average Daily Social Media Usage Hours (Filtered Data)')
        plt.xlabel('Country')
        plt.ylabel('Average Daily Usage (Hours)')
        plt.xticks(rotation=45, ha='right') # Rotate labels to prevent overlap
        plt.tight_layout() # Adjust layout to prevent labels from being cut off
        plt.show()

    else:
        print("df_filtered is empty. Cannot perform analysis on countries with multiple participants.")
elif 'df_filtered' not in locals():
     print("df_filtered DataFrame does not exist. Please run the filtering step first.")
else:
     print("df_filtered is None. Filtering was unsuccessful.")
```

## Task 3 - Question Investigation <a id='task3'></a>

### How are sleep hours , mental health affects their studies and those whose social media use does not?

We will investigate the distributions of `Sleep_Hours_Per_Night`, `Mental_Health_Score` for two groups of students: those who answered 'Yes' to `Affects_Academic_Performance` and those who answered 'No'. 
we will investigate if there is a relationship between a student's relationship status (`Relationship_Status`) and their social media addiction score (`Addicted_Score`). This could reveal if students in different relationship statuses exhibit different levels of social media addiction according to this dataset.

Here we have use boxplot visualization for comparing the distributions of a numerical varible acorss more categories . We will use the original `df_all` for this anlysis.

```python
if df_all is not None:
    # --- Check value counts ---
    print("Counts for Affects_Academic_Performance:")
    print(df_all['Affects_Academic_Performance'].value_counts(dropna=False))

    # --- Define a helper function for boxplot analysis ---
    def boxplot_analysis(df, x, y, title, xlabel, ylabel):
        plt.figure(figsize=(8, 6))
        sns.boxplot(data=df, x=x, y=y,hue=x, palette='deep',legend=False)
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.grid(axis='y', linestyle='--', alpha=0.5)
        plt.tight_layout()
        plt.show()

        print(f"\nDescriptive Statistics for {ylabel} by {xlabel}:")
        print(df.groupby(x)[y].describe())
        print("-" * 40)

    #Analysis 1: Sleep Hours
    boxplot_analysis(
        df_all,
        x='Affects_Academic_Performance',
        y='Sleep_Hours_Per_Night',
        title='Sleep Hours Per Night by Academic Performance Impact',
        xlabel='Affects Academic Performance',
        ylabel='Sleep Hours Per Night'
    )

    #Analysis 2: Mental Health Score
    boxplot_analysis(
        df_all,
        x='Affects_Academic_Performance',
        y='Mental_Health_Score',
        title='Mental Health Score by Academic Performance Impact',
        xlabel='Affects Academic Performance',
        ylabel='Mental Health Score'
    )


    # --- Analysis & Visualization: Addiction Score by Relationship Status ---

    # 1. Create the boxplot
    plt.figure(figsize=(8, 6)) # Adjust figure size

    # Use seaborn boxplot to compare Addiction_Score across Relationship_Status categories
    sns.boxplot(data=df_all, x='Relationship_Status', y='Addicted_Score',hue='Relationship_Status', palette='viridis',legend=False)

    plt.title('Addiction Score Distribution by Relationship Status')
    plt.xlabel('Relationship Status')
    plt.ylabel('Addiction Score')
    plt.show()

    # 2. Print descriptive statistics for Addiction Score per Relationship Status
    print("\nDescriptive Statistics for Addiction Score by Relationship Status:")
    addiction_by_relationship_describe = df_all.groupby('Relationship_Status')['Addicted_Score'].describe()
    print(addiction_by_relationship_describe)

else:
    print("df_all DataFrame does not exist. Please load the dataset first.")
```
